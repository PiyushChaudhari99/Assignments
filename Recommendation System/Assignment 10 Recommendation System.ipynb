## Recommendation System

```python
x = df.groupby('User.ID').count()['Book.Rating'] > 30
users = x[x].index
```

```python
filtered_rating = df[df['User.ID'].isin(users)]
filtered_rating
```

```python
 filtered_rating.groupby('Book.Title').count()['Book.Rating'].sort_values(ascending = False)
```

```python
final_ratings = filtered_rating
```

```python
final_ratings = final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book.Title',columns='User.ID',values='Book.Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
pt1 = df.pivot_table(index='Book.Title',columns='User.ID',values='Book.Rating')
```

```python
pt1.fillna(0,inplace=True)
```

```python
pt1
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt1)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt1.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = df[df['Book.Title'] == pt1.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book.Title')['Book.Title'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('Airframe')
```
